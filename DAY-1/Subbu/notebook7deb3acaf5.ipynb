**MathplotLib Examples**


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0.1,10,100) # create an array with 100 elements equally spaced data between 0.1 to 10
print(x)
```

```python
y1 = np.sin(x) # Sin of x data
y2 = np.cos(x) # Cosine of x data
print(y1)
print(y2)
```

```python
plt.figure(figsize=(8,6)) # plot size 
plt.plot(x,y1,label='sin(x)',color='red', linestyle = '--')
plt.plot(x,y2,label='cos(x)',color='blue', linestyle = '-')
plt.title('Sine and Cosine Function Chart')
plt.xlabel('X-Data')
plt.ylabel('Y-Data')
plt.xlim(0,15)
plt.ylim(-1.5,1.5)
plt.legend()
plt.show()
```

```python
data=data.drop(['S.No.'], axis=1)
data.info()
```

```python
from datetime import date
```

```python
date.today().year
data['Car_Age'] = date.today().year - data['Year']
data.info()
data.head()
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
data['Model'] = data.Name.str.split().str.get(1)+data.Name.str.split().str.get(2)
data.head()
```

```python
data[['Name','Brand','Model']]
```

```python
print(data.Brand.unique())
```

```python
searchfor = ['Mini','Land','Isuzu','ISUZU']
data[data.Brand.str.contains('|'.join(searchfor))].head()
```

```python
data["Brand"].replace({"ISUZU":"Isuzu", "Mini": "Mini Cooper", "Land": "Land Rover"}, inplace=True)
searchfor = ['Mini Cooper','Land Rover','Isuzu']
data[data.Brand.str.contains('|'.join(searchfor))].head()
```

```python
data.describe().T
```

```python
data.describe(include='all')
```

```python
cat_cols = data.select_dtypes(include=['object']).columns
print(cat_cols)
```

```python
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew:',round(data[col].skew(),2))
    
    plt.figure(figsize=(15,4)) # plot size 
    plt.subplot(1,2,1)
    data[col].hist(grid=False)
    plt.xlabel(col)
    plt.ylabel('count')
    
    plt.subplot(1,2,2)
    sns.boxplot(x=data[col])
    plt.xlabel(col)
    plt.ylabel('count')
    
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'Fuel_Type', data = data, color = 'orange', order = data['Fuel_Type'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'Transmission', data = data, color = 'red', order = data['Transmission'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Owner_Type', data = data, color = 'blue', order = data['Owner_Type'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'Location', data = data, color = 'green',
order = data['Location'].value_counts().index);
sns.countplot(ax = axes [2, 0], x = 'Brand', data = data, color = 'black', order = data['Brand' ].head (25).value_counts().index);
sns.countplot(ax = axes [2, 1], x = 'Model', data = data, color = 'grey', order = data['Model' ].head (25).value_counts().index);
axes[1][1].tick_params (labelrotation=45);
axes[2][0].tick_params (labelrotation=90);
axes[2][1].tick_params (labelrotation=90);
```
