```python
columns = ["Name", "Market Cap (US$ Billion)"]

def extract():
    extracted_data = pd.DataFrame(columns=columns)
    for jsonfile in glob.glob("/kaggle/input/casptone-day4-data/*.json"):
        extracted_data = extracted_data._append(extract_from_json(jsonfile), ignore_index=True)
    
    return extracted_data
```

```python
my_extracted_df = extract()
my_extracted_df
```

```python
dataframe = pd.read_csv('/kaggle/input/casptone-day4-data/exchange_rates.csv', index_col=0)
dataframe.head(10)
```

```python
def read_csv(file_to_process):
    dataframe = pd.read_csv(file_to_process, index_col=0)
    return dataframe

def extract_csv():
    for csvfile in glob.glob('/kaggle/input/casptone-day4-data/*.csv'):
        extracted_csv_data = pd.DataFrame(read_csv(csvfile))
    return extracted_csv_data
```

```python
my_csv_extracted = extract_csv()
my_csv_extracted.head(10)
```

```python
exchane_rate = float(my_csv_extracted.loc["GBP"][0])
exchane_rate
```

```python
def transform(data):
    data["Market Cap (US$ Billion)"] = round(data["Market Cap (US$ Billion)"] * exchane_rate, 3)
    data.rename(columns = {"Name": "Name", "Market Cap (US$ Billion)" : "Market Cap (GBP Billion)"}, inplace=True, errors= "raised")
    return data
```

```python
my_transform_df = transform(my_extracted_df)
my_transform_df.head(10)
```

```python
filename = "bank_market_cap_in_gbp.csv"

def load(filename, data):
    data.to_csv(filename, index=False)
```

```python
logfile = "bmc_log.txt"

def log(message):
    timestamp = datetime.now().strftime("Date: %h %d, %Y - Time: %H:%M:%S")
    with open(logfile, "a") as textfile:
        textfile.write(f"{timestamp}:{message}.\n")
```

```python
log("ETL Job Started")
log("Extract Phase Started")
```

```python
extractedData = extract()
extractedData.head(5)
```

```python
log("Extract Phase Ended")
log("Transform Phase Started")
```

```python
transformData = transform(extractedData)
transformData.head(5)
```

```python
log("Transform Phase Ended")
log("Load Phase Started")
```

```python
load(filename, transformData)
```
