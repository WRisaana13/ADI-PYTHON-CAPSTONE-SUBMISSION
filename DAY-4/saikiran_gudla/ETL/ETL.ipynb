```python
def extract_from_json(file_to_process):
    dataframe = pd.read_json(file_to_process)
    return dataframe
```

```python
columns = ["Name", "Market Cap (US$ Billion)"]

# Combine the data of all json files into a single file
def extract():
    extracted_data = pd.DataFrame(columns = columns) #Blank dataframe
    for jsonfile in glob.glob("*.json"): #this var will contain json file name available in the folder one by one
        #Append to the blank df created
        extracted_data = extracted_data._append(extract_from_json(jsonfile), ignore_index=True)

    return extracted_data
```

```python
my_extracted_df = extract()
my_extracted_df.tail(10)
```

```python
def read_csv(file_to_process):
    dataframe = pd.read_csv(file_to_process, index_col=0)
    return dataframe

def extract_csv():
    for csvfile in glob.glob("*.csv"):
        extracted_csv_data = pd.DataFrame(read_csv(csvfile))
    return extracted_csv_data
    
```

```python
my_csv_extracted = extract_csv()
my_csv_extracted
```

```python
rate = float(my_csv_extracted.loc["GBP"])
rate
```

```python
def transform(data):
    data["Market Cap (US$ Billion)"] = round(data["Market Cap (US$ Billion)"] * rate, 3)
    data.rename(columns = {"Name": "Name", "Market Cap (US$ Billion)" : "Market Cap (GBP Billion)"}, inplace =True, errors = "raised")
    return data
```

```python
my_transform_df = transform(my_extracted_df)
my_transform_df.head(5)
```

```python
filename = "bank_market_cap_gbp.csv"

def load(filename, data):
    data.to_csv(filename, index=False)
```

```python
import time
logfilename = "bmc_log.txt"

def log(message):
    timestamp = datetime.now().strftime("Date: %h %d, %Y - Time: %H:%M:%S")
    with open(logfilename, "a") as txtfile:
        txtfile.write(f"{timestamp} : {message}. \n")
```

```python
log("ETL Job Started")
time.sleep(1)
log("Extract Phase started")
```

```python
extractedData = extract()
extractedData.head(5)
```

```python
log("Extract Phase Ended")
time.sleep(1)
log("Transform Phase Started")
```

```python
transformedData = transform(extractedData)
transformedData.head(5)
```

```python
log("Transform Phase Ended")
time.sleep(1)
log("Load Phase Started")
```
